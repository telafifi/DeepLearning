```python
import matplotlib.pyplot as plt
import torch
```

```python
# function defined
def f(x: torch.Tensor) -> torch.Tensor:
    return 3*x**2 + torch.sin(5*x)
```

```python
x_base = torch.arange(-1, 1, 0.01)
plt.plot(x_base, f(x_base))
```

```python
# Need to compute the gradient of the function
def grad_f(x: torch.Tensor) -> torch.Tensor:
  return 6*x + 5*torch.cos(5*x)
```

```python
# Implement gradient descent, starting with a random number
x = torch.randn(1)

all_x = []

for _ in range(100):
  x = x - 0.01 * grad_f(x)
  all_x.append(x)
  
plt.plot(x_base, f(x_base))
plt.plot(torch.stack(all_x), f(torch.stack(all_x)), 'ro')

# We will see that given the random number, we can sometimes get stuck in a local minimum
```
